```python
filenames=[]
for year in [2013,2014,2015]:
    year=str(year)
    for xx in range(1,13):
        x=str(xx)
        if len(x) <2:
            x='0'+x

        filenames.append('https://database.lichess.org/standard/lichess_db_standard_rated_'+year+'-'+x+'.pgn.bz2'+'\n')
filenames
file1 = open('download_chess.txt', 'w')
# Writing multiple strings
# at a time
file1.writelines(filenames)
  
# Closing file
file1.close()
```
